```python
offers_dict = {
    "null": 0.0,
    "offer_1": 0.05,
    "offer_2": 0.1,
    "offer_3": 0.15,
    "offer_4": 0.20,
    "offer_5": 0.20
}
```

```python
from typing import Dict

def map_offers_to_discount_pandas(df: pd.DataFrame, offers_dict: Dict[str, float]) -> pd.DataFrame:
    def calculate_discount(row):
        for offer_col in ["offer_1", "offer_2", "offer_3", "offer_4", "offer_5"]:
            if row[offer_col] == True:
                return offers_dict[offer_col]
        return offers_dict["null"]
    
    df['discount'] = df.apply(calculate_discount, axis=1)
    return df


new_df = map_offers_to_discount_pandas(df, offers_dict)
```
